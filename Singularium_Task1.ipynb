**Importing Dependencies**

```python
import pandas as pd
import json
from collections import defaultdict
```

```python
df=pd.read_csv("classification.csv")
df
```

```python
columns_to_drop = ['meta_row_data', 'verification_messages','order', 'jpg_url','meta_verification_messages', 'meta_classification_1',
       'meta_classification_2', 'meta_classification_3', 'meta_order',
       'meta_jpg_url', 'meta_file_name']

df1 = df.drop(columns=columns_to_drop)

# Save the updated CSV file
output_csv = "/content/updated_classification.csv"
df1.to_csv(output_csv, index=False)
```

```python
df1
```

```python
# Load JSON file (replace 'file.json' with your file path)
with open('tag_outputs.json', 'r') as f:
    json_data = json.load(f)

```

```python
json_df = pd.DataFrame(json_data)

json_df
```

```python
data = df1.merge(json_df, how='left', left_on='file_name', right_on='key')
data
```

```python
data = data.drop(index=0)

data = data.reset_index(drop=True)
data
```

```python
data["sub_class"] = data["sub_class"].replace('', 'IGNORE')
data
```

```python
accuracy_results = defaultdict(list)
total_predictions = len(data)
correct_predictions = 0
```

```python
import numpy as np
for index, row in data.iterrows():
    subclass = row.get('sub_class', 'Unknown')
    classifications = [
        row.get('classification_1', ''),
        row.get('classification_2', ''),
        row.get('classification_3', ''),
    ]

    # Exceptional cases for accurate predictions
    if pd.isna(subclass) or subclass in ['CBC|PATHOLOGY', 'IGNORE']:
        correct_predictions += 1
        accuracy_results[subclass].append(True)
    elif subclass in classifications:
        correct_predictions += 1
        accuracy_results[subclass].append(True)
    else:
        accuracy_results[subclass].append(False)
```

```python
if total_predictions > 0:
    overall_accuracy = (correct_predictions / total_predictions) * 100
else:
    overall_accuracy = 0

print(f"Overall Accuracy: {overall_accuracy:.2f}%")
```

```python
# unique file processing per day

file_processing_per_day = json_df['key'].nunique()
```

```python
accuracy_by_sub_class = {subclass: (sum(matches) / len(matches)) * 100 for subclass, matches in accuracy_results.items()}
accuracy_df = pd.DataFrame(list(accuracy_by_sub_class.items()), columns=['Sub-Class', 'Accuracy (%)'])
print("Accuracy by Sub-Class:\n", accuracy_df.to_string(index=False))
```

```python
#results for reporting
report = {
    'accuracy_by_sub_class': {subclass: (sum(matches) / len(matches)) * 100 for subclass, matches in accuracy_results.items()},
    'overall_accuracy': f"{overall_accuracy:.2f}%",
    'Pages_processing_per_day': file_processing_per_day,
}
```

```python
import csv
from datetime import datetime
output_file = "report.csv"
current_date = datetime.now().strftime("%Y-%m-%d")

# Flatten and write the report data to a CSV file
with open(output_file, mode="w", newline="") as file:
    writer = csv.writer(file)

    writer.writerow(["Report Generation Date", current_date])
    writer.writerow([])

    # Write overall accuracy
    writer.writerow(["Metric", "Value"])
    writer.writerow(["Overall Accuracy", report['overall_accuracy']])
    writer.writerow(["Total Pages Processed Per Day", report['Pages_processing_per_day']])
    writer.writerow([])  # Empty row for spacing

    # Write accuracy by sub-class
    writer.writerow(["Sub Class", "Accuracy (%)"])
    for subclass, accuracy in report['accuracy_by_sub_class'].items():
        writer.writerow([subclass, accuracy])

print(f"Report successfully written to {output_file}")
```

```python
import csv

# Function to generate HTML page from CSV data
def generate_html(csv_file, output_html):
    with open(csv_file, mode='r') as file:
        reader = csv.reader(file)

        # Extract all rows
        rows = list(reader)

    # Start HTML structure
    html_content = '''
    <!DOCTYPE html>
    <html lang="en">
    <head>
        <meta charset="UTF-8">
        <meta name="viewport" content="width=device-width, initial-scale=1.0">
        <title>Report Summary</title>
        <style>
            body {
                font-family: Arial, sans-serif;
                margin: 20px;
                background-color: #f4f4f4;
            }
            h1 {
                text-align: center;
            }
            table {
                width: 50%;
                margin: 20px auto;
                border-collapse: collapse;
            }
            th, td {
                padding: 5px;
                border: 1px solid #ccc;
                text-align: left;
            }
            th {
                background-color: #f2f2f2;
            }
            .section-title {
                font-size: 1.5em;
                margin-top: 40px;
                text-align: center;
            }
            .logo {
                position: absolute;
                top: 10px;  /* Adjust to position from the top */
                right: 10px;  /* Adjust to position from the right */
                width: 60px;  /* Smaller logo size */
            }
        </style>
    </head>
    <body>

        <img src="logo.png" alt="Company Logo" class="logo">

        <h1>Report Summary</h1>
    '''

    # Process CSV data for Report Generation Date
    report_generation_date = rows[0][1]
    html_content += f'''
    <table>
        <tr>
            <th>Report Generation Date</th>
            <td>{report_generation_date}</td>
        </tr>
    </table>
    '''

    # Process Overall Metrics
    html_content += '<div class="section-title">Overall Metrics</div><table><tr><th>Metric</th><th>Value</th></tr>'
    for row in rows[3:5]:
        html_content += f'<tr><td>{row[0]}</td><td>{row[1]}</td></tr>'
    html_content += '</table>'

    # Process Sub Class Accuracy
    html_content += '<div class="section-title">Sub Class Accuracy</div><table><tr><th>Sub Class</th><th>Accuracy (%)</th></tr>'
    for row in rows[7:]:
        if row:
            accuracy_value = float(row[1])
            html_content += f'<tr><td>{row[0]}</td><td>{accuracy_value:.2f}%</td></tr>'
    html_content += '</table>'

    # End HTML structure
    html_content += '''
    </body>
    </html>
    '''

    # Write the HTML content to a file
    with open(output_html, 'w') as output_file:
        output_file.write(html_content)

# Call the function to generate HTML
generate_html('report.csv', 'output_report.html')
```
